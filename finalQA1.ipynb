```python
import pandas as pd
import numpy as np
```

```python
# Reading .csv file as Numpy Ndarray
dataset = pd.read_csv("diabetes.csv", header=None).values 
```

```python
# print the shape of the dataset
print(type(dataset))
print(dataset.shape)
```

```python
from sklearn.model_selection import train_test_split

# Last column is our the target which is 8 th column
# used the split funciton to split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8], test_size=0.25, random_state=42)

print("X_train.shape: ", X_train.shape)
print("Y_train.shape: ", Y_train.shape)
print("X_test.shape: ", X_test.shape)
print("Y_test.shape: ", Y_test.shape)
```

```python
from keras.models import Sequential
import tensorflow as tf
from tensorflow import keras
# creating the Sequential model
my_first_nn = Sequential() 

# FIX ME
# defined the Input Layer shape 
my_first_nn.add(keras.Input(shape=(8,))) 

# added 32 layer 
my_first_nn.add(keras.layers.Dense(32, activation='relu')) # hidden layer

my_first_nn.add(keras.layers.Dense(1, activation='sigmoid'))

my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
# printed the summary of the compilation of the neural network
print(my_first_nn.summary())

# The returned history object holds a record of the loss values and metric values during training
history =  my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100, verbose=1, shuffle=True, validation_data=(X_test,Y_test))
```

```python
from matplotlib import pyplot as plt
# plotted the test and train accurancy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Training Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
# predicted using the model and evaluated it
print(my_first_nn.evaluate(X_test, Y_test))
```

```python
from matplotlib import pyplot as plt
# plotted the test and train accurancy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Training Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()
```
